```python
## Need to get and load each item contained in the given STAC catalogue
## Catalogue url
url='https://api.stac.ceda.ac.uk/collections/sentinel2_ard/items'

response = requests.get(url)

json_out = (response.json())

json_features = json_out['features']
json_features
```

```python
## Load each feature into the resource catalogue
for feature in json_features:
    response = requests.post(
        f"{registration_endpoint}/register-json",
        json=feature,
    verify=False
    )
    #response.raise_for_status()
    print(response)
    break
    # if response != "<Response [202]":
    #     print("Item not added!")
```
